```python
# import data
import pandas as pd

apps = pd.read_csv('datasets/apps.csv')
apps.info()
apps.head()
```

```python
# Q1 remove , and + from data in Installs by first defining the characters to remove
chars_to_remove = ['+', ',']

# Loop for each char in chars_to_remove
for char in chars_to_remove:
   # Replace the character with an empty string
    apps['Installs'] = apps['Installs'].apply(lambda x: x.replace(char, ''))
    # Convert col to float data type
    
apps['Installs'] = apps['Installs'].astype(int)
apps.info()
apps.sample(10)
```

```python
# Q2 first extract mean for each column than append app count per category as a new column
app_category_info = apps.groupby('Category').mean()
app_category_info['Number of apps'] = apps.groupby('Category').count()['App']
app_category_info.reset_index(level=0, inplace=True)

# Extract the require columns for the question and rename them to the names specified in the question
app_category_info = app_category_info.loc[:, ['Category', 'Number of apps', 'Price', 'Rating']]
app_category_info = app_category_info.rename(columns={'Price': 'Average price', 'Rating': 'Average rating'})

app_category_info
```

```python
# Q3 import user_reviews.csv then merge it with the apps dataframe using inner join
reviews_df = pd.read_csv('datasets/user_reviews.csv')
merged_df = apps.merge(reviews_df, on = 'App', how = 'inner')

# subset merged data so only finance category and free apps remain
finance_df = merged_df.loc[(merged_df.Category == 'FINANCE') & (merged_df.Type == 'Free'),]

# Compute average sentiment score for each app and order them in descending order then subset rows where SC is higher than average
app_SC_avg = finance_df.groupby('App').agg({'Sentiment Score':'mean'})\
                       .sort_values('Sentiment Score', ascending = False)
top_finance = app_SC_avg.loc[app_SC_avg['Sentiment Score'] > app_SC_avg['Sentiment Score'].mean(),]

# extract top finance apps (e.g., those higher than the mean) and sort them in desecnding order
top_10_user_feedback = top_finance.iloc[0:10]
top_10_user_feedback
```
